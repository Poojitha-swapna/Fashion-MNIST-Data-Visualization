```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Fashion-MNIST dataset
from tensorflow.keras.datasets import fashion_mnist

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Print dataset shape
print("Training data shape:", x_train.shape)
print("Training labels shape:", y_train.shape)
```

```python
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
```

```python
# Display one sample image
plt.figure()
plt.imshow(x_train[0], cmap='gray')
plt.title(class_names[y_train[0]])
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(8,8))

for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')

plt.show()
```

```python
import pandas as pd

# Count number of images per class
label_counts = pd.Series(y_train).value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.barplot(x=class_names, y=label_counts.values)
plt.xticks(rotation=45)
plt.title("Class Distribution in Fashion-MNIST Dataset")
plt.ylabel("Number of Images")
plt.xlabel("Clothing Category")
plt.show()
```

```python
plt.figure(figsize=(12,6))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    avg_image = np.mean(x_train[y_train == i], axis=0)
    plt.imshow(avg_image, cmap='gray')
    plt.title(class_names[i])
    plt.axis('off')

plt.suptitle("Average Image for Each Fashion-MNIST Class", fontsize=14)
plt.show()
```

```python
# Flatten images to 1D to see correlation
x_train_flat = x_train.reshape(x_train.shape[0], -1)

# Compute correlation between first 100 images
corr = np.corrcoef(x_train_flat[:100])

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap of Sample Images")
plt.show()
```
